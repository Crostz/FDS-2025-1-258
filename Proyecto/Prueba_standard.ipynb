```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("USvideos.csv")
```

```python
df_n = df.fillna({"description":df["title"]}) # Llenamos valores faltantes.
df = df_n
```

```python
df_a = df[["views", "dislikes", "likes", "comment_count", "lat", "lon"]]
df_b = df[["views", "dislikes", "likes", "comment_count", "lat", "lon"]]
```

```python
df_b = df[["views", "dislikes", "likes", "comment_count", "lat", "lon"]]
# Al haber una gran distribución desigual, se aplicará una transformación
df_b["likes"] = np.log(df_b["likes"] + 1)  # Se agrega 1 para evitar log(0)
np.expm1(df_b["likes"])  # Inversa de logaritmo natural
df_b["dislikes"] = np.log(df_b["dislikes"] + 1)  # Se agrega 1 para evitar log(0)
np.expm1(df_b["dislikes"])  # Inversa de logaritmo natural
df_b["views"] = np.log(df_b["views"] + 1)  # Se agrega 1 para evitar log(0)
np.expm1(df_b["views"])  # Inversa de logaritmo natural
df_b["comment_count"] = np.log(df_b["comment_count"] + 1)  # Se agrega 1 para evitar log(0)
np.expm1(df_b["comment_count"])  # Inversa de logaritmo natural
df_b["lat"] = np.log(df_b["lat"] + 1)  # Se agrega 1 para evitar log(0)
np.expm1(df_b["lat"])  # Inversa de logaritmo natural
df_b["lon"] = np.log(df_b["lon"] + 1)  # Se agrega 1 para evitar log(0)
np.expm1(df_b["lon"])  # Inversa de logaritmo natural
```
