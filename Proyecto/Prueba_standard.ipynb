```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
# df_dt = pd.DataFrame(df["publish_time"])
# df_dt = df_dt.assign(Hour = pd.DatetimeIndex(df["publish_time"]).hour)
# df_dt = df_dt.assign(Day = pd.DatetimeIndex(df["publish_time"]).day)
df_b = df_b.assign(Hour = pd.DatetimeIndex(df["publish_time"]).hour)
df_b = df_b.assign(Day = pd.DatetimeIndex(df["publish_time"]).day)

df_b["Day"] = np.log10(df_b["Day"] + 1) 
np.expm1(df_b["Day"]) 
df_b["Day"] = np.log(df_b["Day"] + 1)  # Se agrega 1 para evitar log(0)
np.expm1(df_b["Day"])  # Inversa de logaritmo natural
df_b["Hour"] = np.log(df_b["Hour"] + 1)  # Se agrega 1 para evitar log(0)
np.expm1(df_b["Hour"])  # Inversa de logaritmo natural
```

```python
X = df_b[['views', 'comment_count', 'dislikes', "Hour", "Day"]]
y = df_b['likes']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
regr = RandomForestRegressor(max_depth=10, random_state=101)
regr.fit(X_train, y_train)
```

```python
Y_pred = regr.predict(X_test)
pred3 = pd.DataFrame(X_test)
pred3 = pred3.assign(Likes=y_test)
pred3 = pred3.assign(Pred = Y_pred.flatten().tolist())
pred3.head(7)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, Y_pred))
print('MSE:', metrics.mean_squared_error(y_test, Y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, Y_pred)))
```
